# Scrape Cocktails

```python
import re
import json
import requests
url = 'https://www.diffordsguide.com/sitemap/gb.xml'
sitemap = requests.get(url)
urls = re.findall('<loc>(https://www.diffordsguide.com/cocktails/recipe/.{3,50})</loc>', str(sitemap.content))
```

```python
%%time
for ix, url in enumerate(urls):
    print('PULLING URL {}:'.format(ix), url)
    content = str(requests.get(url).content)
    recipe_id = url.split('recipe/')[1].split('/')[0]
    with open('./diffordsguide_cocktails/' + recipe_id + '.txt', 'w') as f:
        f.write(content)
```

# Scrape Ingredients 

### Requires cocktails_db.pkl generation

```python
import pandas as pd
import pickle
cocktails = pickle.load(open('../2_data_format_and_export/cocktails_db.pkl', 'rb'))
df = pd.DataFrame([ing for ing_list in cocktails.ingredients for ing in ing_list])
print('-----> Found {} total ingredients for {} total cocktails'.format(len(df), len(cocktails)))
df = df.drop_duplicates('name')
print('-----> Found {} unique ingredients de-duping on name'.format(len(df)))
```

```python
sdf = df.loc[pd.notnull(df.link)]
print('-----> Scraping {} total urls'.format(len(sdf)))
for ix, link in enumerate(sdf.link):
    url = 'https://diffordsguide.com' + link
    print('PULLING URL {}:'.format(ix), url)
    content = str(requests.get(url).content)
    ingredient_id = link.replace('/','_')
    with open('./diffordsguide_ingredients/' + ingredient_id + '.txt', 'w') as f:
        f.write(content)
```
